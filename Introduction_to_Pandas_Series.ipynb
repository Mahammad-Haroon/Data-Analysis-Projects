# Smart Energy Monitoring System
## Using Pandas Series

```python
import pandas as pd
```

```python
# Energy consumption values for House A
energy_A = [12.5, 13.2, 11.8, 14.0, 15.1]

# Custom day labels
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']

# Create Pandas Series
house_A = pd.Series(energy_A, index=days)

# Display the Series
print("House A Energy Consumption:")
print(house_A)
```

Access Wednesday Value

```python
# Retrieve Wednesday energy consumption
wed_energy = house_A.loc['Wed']

# Print only that value
print("Energy consumption on Wednesday:", wed_energy)
```

Create Series from Dictionary

```python
# Dictionary for House B
energy_B_dict = {
    'Mon': 10.2,
    'Tue': 11.5,
    'Thu': 12.0,
    'Fri': 13.3
}

# Create Series
house_B = pd.Series(energy_B_dict)

# Display Series
print("House B Energy Consumption:")
print(house_B)
```

Add the Two Series

```python
# Add both Series
total_consumption = house_A + house_B

# Display result
print("Total Energy Consumption (House A + House B):")
print(total_consumption)
```
