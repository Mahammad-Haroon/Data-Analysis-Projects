```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.info()
```

Task 2 – Type Conversion using .astype()

```python
df["total_bill"] = df["total_bill"].astype("float64")
df["tip"] = df["tip"].astype("float64")
```

```python
df["size"] = df["size"].astype("int64")
```

```python
df["day"] = df["day"].astype("object")
```

```python
df.dtypes
```

Task 3 – Semantic Transformation

```python
df["sex"] = df["sex"].map({"Male": "M", "Female": "F"})
```

```python
df["smoker"] = df["smoker"].map({"Yes": True, "No": False})
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

Task 4 – String Cleaning

```python
df["time"] = df["time"].str.lower()
```

```python
df["time"] = df["time"].str.strip()
```
