```python
import pandas as pd
import numpy as np

data = {
    'Customer_ID': [1, 2, 3, 4, 5, 6, 8, 9, 10],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan'],
    'Age': [25, np.nan, 30, 22, 45, np.nan, 33, 38, np.nan],
    'City': ['New York', 'London', np.nan, 'London', 'Paris', 'New York', 'London', np.nan, 'Paris'],
    'Spend_Amount': [150, 200, 120, np.nan, 300, 180, np.nan, 210, 190],
    'Feedback_Notes': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
}

df = pd.DataFrame(data)

df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Cleaning Strategy:

1. The 'Feedback_Notes' column will be dropped entirely because it contains 100% missing values and provides no useful information.

2. For the 'Age' column, I will use the Mean to fill missing values. There are no extreme outliers in the dataset, so the mean provides a fair estimate.

3. For the 'City' column, I will use the Mode (most frequent city) to fill missing values because it is categorical data.

```python
df.dropna(axis=1, how='all', inplace=True)
```

Fill Age with Mean

```python
age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
```

```python
age_mean
```

Fill City with Mode

```python
city_mode = df['City'].mode()[0]
df['City'] = df['City'].fillna(city_mode)
```

Drop Rows Where Spend_Amount is Missing

```python
df.dropna(subset=['Spend_Amount'], inplace=True)
```

Final Null Check

```python
df.isnull().sum()
```
