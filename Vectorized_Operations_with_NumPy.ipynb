```python
import numpy as np
```

```python
# Task 1: Initialize Energy Consumption Data

energy_data = np.array([
    [12, 14, 13, 15, 16, 14],
    [10, 11, 12, 13, 14, 12],
    [20, 22, 21, 23, 24, 22],
    [8, 9, 10, 11, 12, 10],
    [15, 16, 17, 18, 19, 17]
])

# Print the array
print("Energy Data (kWh):")
print(energy_data)

# Print shape
print("\nShape of energy data:", energy_data.shape)
```

```python
# Task 2: Apply Load Adjustment Using Vectorization

adjusted_energy = energy_data * 1.10

print("Adjusted Energy Consumption (+10%):")
print(adjusted_energy)
```

```python
# Task 3: Hour-Based Calibration Using Broadcasting

calibration_factors = np.array([1.0, 0.98, 1.02, 1.01, 0.99, 1.03])

calibrated_energy = adjusted_energy * calibration_factors

print("Calibrated Energy Consumption:")
print(calibrated_energy)
```

```python
# Average energy per machine

avg_energy_per_machine = np.mean(calibrated_energy, axis=1)

print("Average Energy Consumption per Machine:")
print(avg_energy_per_machine)
```

```python
# Maximum energy value

max_energy = np.max(calibrated_energy)

print("\nMaximum Energy Consumption in Dataset:")
print(max_energy)
```
