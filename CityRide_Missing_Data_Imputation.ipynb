# CityRide Data Cleaning and Imputation
### Missing Data Strategy and Impact Analysis

## Step 0: Generate Dataset

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

df = pd.DataFrame({
    "Ride_ID": [f"RIDE_{i+1}" for i in range(n)],
    "Ride_Distance": np.random.normal(8, 2.5, n).round(2),
    "Fare_Amount": np.random.normal(250, 80, n).round(2),
    "Driver_Rating": np.random.choice([3, 4, 5, np.nan], size=n, p=[0.2, 0.2, 0.12, 0.48]),
    "Vehicle_Type": np.random.choice(["Sedan", "SUV", "Mini", np.nan], size=n, p=[0.3, 0.3, 0.34, 0.06]),
    "Outside_Temperature": np.random.normal(30, 4, n).round(1),
    "Ride_Time": pd.date_range(start="2023-01-01", periods=n, freq="H")
})

# Introduce additional missingness
df.loc[df.sample(frac=0.03).index, "Ride_Distance"] = np.nan
df.loc[df.sample(frac=0.01).index, "Fare_Amount"] = np.nan
df.loc[df.sample(frac=0.12).index, "Outside_Temperature"] = np.nan

df.head()
```

## Task 1: Missing Data Assessment

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent
```

```python
strategy_table = pd.DataFrame({
    "Column": df.columns,
    "Data_Type": df.dtypes.values,
    "Percent_Missing": missing_percent.values,
    "Chosen_Strategy": [
        "None",
        "Median",
        "Mean",
        "Flag_and_Zero",
        "Mode",
        "Forward_Fill",
        "None"
    ]
})

strategy_table
```

## Task 2: Apply Imputation Strategies

```python
original_df = df.copy()
```

```python
df["Ride_Distance"].fillna(df["Ride_Distance"].median(), inplace=True)
```

```python
df["Fare_Amount"].fillna(df["Fare_Amount"].mean(), inplace=True)
```

```python
df["Vehicle_Type"].fillna(df["Vehicle_Type"].mode()[0], inplace=True)
```

```python
df = df.sort_values("Ride_Time")
```

```python
df["Outside_Temperature"].fillna(method="ffill", inplace=True)
```

```python
df["Outside_Temperature"].fillna(method="ffill", inplace=True)
```

```python
df["Driver_Rating"].fillna(0, inplace=True)
```

```python
comparison = pd.DataFrame({
    "Original_Nulls": original_df.isnull().sum(),
    "Cleaned_Nulls": df.isnull().sum()
})

comparison
```

## Task 3: Behavioral Impact Analysis

```python
df["Driver_Rating_Missing"] = df["Driver_Rating"].isnull().astype(int)
```

```python
df["Driver_Rating"].fillna(0, inplace=True)
```

```python
avg_missing = df[df["Driver_Rating_Missing"] == 1]["Fare_Amount"].mean()
avg_missing
```

```python
avg_not_missing = df[df["Driver_Rating_Missing"] == 0]["Fare_Amount"].mean()
avg_not_missing
```

```python
analysis_table = pd.DataFrame({
    "Group": ["Missing Rating", "Non-Missing Rating"],
    "Average_Fare": [avg_missing, avg_not_missing],
    "Ride_Count": [
        df[df["Driver_Rating_Missing"] == 1].shape[0],
        df[df["Driver_Rating_Missing"] == 0].shape[0]
    ]
})

analysis_table
```
