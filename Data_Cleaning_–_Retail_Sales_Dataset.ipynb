**Objective:** Clean a messy retail sales dataset using Pandas for accurate analysis.


Load the Dataset

```python
import pandas as pd
import io

csv_data = """ Order ID , Region ,Category , Sales , Quantity , Order Date
1001 , north , Electronics , "$1,200" , 10 units , 01-01-2024
1002 , South, electronics , "$850" , 5 Units , 02-01-2024
1003 , EAST , Furniture , "$2,500" , 3 units , 03-01-2024
1004 , west , furnture , "$1,750" , 2 units , 04-01-2024
1005 , Noth , Electronics , "$1,200" , 10 units , 01-01-2024
1005 , Noth , Electronics , "$1,200" , 10 units , 01-01-2024
1006 , south , Office Supplies , "$300" , 7 units , 05-01-2024
1007 , East , office supplies , "$450" , 4 Units , 06-01-2024
1008 , WEST , Electronics , "$1,100" , 6 units , 07-01-2024
"""

df = pd.read_csv(io.StringIO(csv_data), skipinitialspace=True)

print("Initial Dataset")
df.head()
```

Inspect Dataset

## Inspecting Dataset
Checking column names, data types, and duplicates.

```python
print("Column Names:")
print(df.columns)

print("\nDataset Info:")
df.info()
```

This shows:

spaces in column names

incorrect data types

## Task 1 – Fix Structural Issues

1. Remove whitespace from column names
2. Convert column names to Title Case
3. Remove duplicate rows

```python
# Remove spaces from column names
df.columns = df.columns.str.strip()

# Convert column names to Title Case
df.columns = df.columns.str.title()

# Remove duplicate rows
df = df.drop_duplicates()

df.head()
```

## Task 2 – Clean Region Column
- Remove spaces
- Convert to Title Case
- Fix typo "Noth"

```python
df['Region'] = df['Region'].str.strip()
df['Region'] = df['Region'].str.title()

# Fix typo
df['Region'] = df['Region'].replace("Noth","North")

df['Region'].value_counts()
```

## Task 3 – Clean Category Column
Standardizing category names and fixing spelling errors.

```python
df['Category'] = df['Category'].str.strip()
df['Category'] = df['Category'].str.title()

# Fix typo
df['Category'] = df['Category'].replace("Furnture","Furniture")

df['Category'].value_counts()
```

## Task 4 – Clean Sales Column
Removing currency symbols and converting to float.

```python
df['Sales'] = df['Sales'].str.replace('$','', regex=False)
df['Sales'] = df['Sales'].str.replace(',','', regex=False)

df['Sales'] = df['Sales'].astype(float)

df[['Sales']].head()
```

## Task 5 – Clean Quantity Column
Removing unit text and converting to integer.

```python
df['Quantity'] = df['Quantity'].str.lower()
df['Quantity'] = df['Quantity'].str.replace('units','', regex=False)

df['Quantity'] = df['Quantity'].astype(int)

df[['Quantity']].head()
```

## Task 6 – Convert Order Date to Datetime

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df[['Order Date']].head()
```
